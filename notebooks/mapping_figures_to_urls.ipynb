```python
def find_deadlinks(script_tuple):
    failed = []
    fig_no = script_tuple[0]
    urls = script_tuple[1]
    for url in urls:
        try:
            resp = requests.get(url)
        except:
            failed.append([fig_no, url, "ConnectionError"])
            continue
        if resp.status_code != 200:
            failed.append([fig_no, url, resp.status_code])
    if failed == []:
        return None
    return failed
```

```python
import pdfminer
import os
import regex as re
import fitz  # this is pymupdf
import requests
```

```python
with fitz.open("pml1-listoffigures.pdf") as doc:
    text = ""
    for page in doc:
        text += page.get_text()

print(text)
```

```python
pattern = r"[\n][0-9]+[.][0-9]+[\n]"
fig_numbers = re.findall(pattern, text)
len(fig_numbers)
```

```python
captions = re.compile(pattern).split(text)[1:]
len(captions)
```

```python
book_no = 1
chapter_no = 2
base_url_py = "https://github.com/probml/pyprobml/blob/master/scripts"
base_url_ipynb = f"https://github.com/probml/pyprobml/blob/master/notebooks/book{book_no}/{chapter_no:02d}"
base_url_ipynb
```

```python
seperator_pattern = r"[\s\n]"
scripts = {}
nb = set()
for no, cap in zip(fig_numbers, captions):
    words = re.compile(seperator_pattern).split(cap)
    fig_no = no.strip()
    #     print("********************************************************")
    #     print(fig_no, cap)

    script_names = []
    for each in words:
        if ".py" in each:
            script_names.append(os.path.join(base_url_py, each[: each.find(".py") + 3]))
        elif ".ipynb" in each:
            script_names.append(os.path.join(base_url_ipynb, each[: each.find(".ipynb") + 6]))
            nb.add(each[: each.find(".ipynb") + 6])
    print(fig_no, script_names)
    scripts[fig_no] = script_names
#     print("********************************************************")
```
