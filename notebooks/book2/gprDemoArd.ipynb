```python
def plot_surface(data, save_name=""):
    figsize = None if LATEXIFY else (12, 7)
    fig, ax = plt.subplots(1, len(data), figsize=figsize, subplot_kw={"projection": "3d"})
    ax = [ax] if len(data) == 1 else ax
    for index, entry in enumerate(data):
        l1, l2 = entry["l1"], entry["l2"]
        kernel = tinygp.transforms.Linear(1.0 / jnp.array([l1, l2]), tinygp.kernels.ExpSquared())
        gp = GaussianProcess(kernel, x)
        y = gp.sample(key).reshape(N, N)

        mappable = ax[index].plot_surface(x1, x2, y, lw=0.5, cmap="viridis")
        ax[index].set_xlabel("input $x_1$", labelpad=-1)
        ax[index].set_zlabel(
            "output $y$",
            labelpad=-1,
        )
        ax[index].view_init(None, 180 + 50)
        ax[index].set_zlim(-1.0, 2.0)

    ax[0].set_ylabel("input $x_2$", labelpad=-1)
    fig.colorbar(mappable)
    if LATEXIFY and len(save_name) > 0:
        savefig(save_name, bbox_inches="tight")  # Use 'tight' only when necessary (3d plots)


def plot_contour(data, save_name=""):
    figsize = None if LATEXIFY else (12, 4)
    fig, ax = plt.subplots(1, len(data), figsize=figsize, sharex=True, sharey=True)
    ax = [ax] if len(data) == 1 else ax
    for index, entry in enumerate(data):
        l1, l2 = entry["l1"], entry["l2"]
        kernel = tinygp.transforms.Linear(1.0 / jnp.array([l1, l2]), tinygp.kernels.ExpSquared())
        gp = GaussianProcess(kernel, x)
        y = gp.sample(key).reshape(N, N)

        mappable = ax[index].contourf(x1, x2, y, cmap="viridis", levels=20)
        ax[index].set_xlabel("input $x_1$")
        ax[index].set_aspect(0.8)

    ax[0].set_ylabel("input $x_2$")
    cbar = fig.colorbar(mappable)
    cbar.ax.get_yaxis().labelpad = 10
    cbar.ax.set_ylabel("output $y$", rotation=270)

    if LATEXIFY and len(save_name) > 0:
        savefig(save_name)
```

```python
r = jnp.arange(-3, 3.1, 0.1)
N = len(r)
x1, x2 = jnp.meshgrid(r, r)
x = jnp.array([(i, j) for i, j in zip(x1.ravel(), x2.ravel())])
key = jax.random.PRNGKey(1)
```

```python
data = [{"l1": 1.0, "l2": 1.0}, {"l1": 1.0, "l2": 5.0}]
plot_surface(data, save_name="gpDemoArd_surface_latexified.pdf")
```

```python
data = [{"l1": 1.0, "l2": 1.0}, {"l1": 1.0, "l2": 5.0}]
plot_contour(data, save_name="gpDemoArd_contour_latexified.pdf")
```
