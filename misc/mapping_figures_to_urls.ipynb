```python
import os
import regex as re

try:
    import fitz  # this is pymupdf
except:
    %pip install fitz
import requests
```

```python
with fitz.open("pml1-listoffigures.pdf") as doc:
    text = ""
    for page in doc:
        text += page.get_text()

# print(text)
```

```python
pattern = r"[\n][0-9]+[.][0-9]+[\n]"
fig_numbers = re.findall(pattern, text)
len(fig_numbers)
```

```python
captions = re.compile(pattern).split(text)[1:]
len(captions)
```

```python
seperator_pattern = r"[\s\n]"
scripts = {}
book_no = 1
for no, cap in zip(fig_numbers, captions):
    words = re.compile(seperator_pattern).split(cap)
    fig_no = no.strip()
    chapter_no = int(fig_no.split(".")[0])
    base_url_ipynb = f"https://github.com/probml/pyprobml/blob/master/notebooks/book{book_no}/{chapter_no:02d}"
    script_names = []

    for each in words:
        if ".py" in each:
            script_name = each[: each.find(".py")]
            script_names.append(os.path.join(base_url_ipynb, script_name + ".ipynb"))
        elif ".ipynb" in each:
            script_name = each[: each.find(".ipynb")]
            script_names.append(os.path.join(base_url_ipynb, script_name + ".ipynb"))
    print(fig_no, script_names)
    scripts[fig_no] = script_names
```

```python
more_than_one = 0
for each in scripts:
    if len(scripts[each]) > 1:
        print(each, scripts[each])
        more_than_one += 1
more_than_one
```

```python
cnt = 0
for fig in scripts:
    if scripts[fig] != []:
        cnt += 1
cnt
```

```python
dead_urls = []
for fig in scripts:
    for url in scripts[fig]:
        resp = requests.get(url)
        if resp.status_code != 200:
            print(fig, url)
            dead_urls.append([fig, url])
```
