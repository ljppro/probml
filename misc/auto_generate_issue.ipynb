```python
import os
from glob import glob
import pandas as pd
```

```python
base_url = "https://github.com/probml/pyprobml/tree/master/"
metadata = pd.read_pickle("../notebooks/metadata_book1.pkl")
```

```python
right = "&#9989;"
wrong = "&#10060;"
repo_root = "https://github.com/probml/pyprobml"
run_root = "https://raw.githubusercontent.com/probml/pyprobml/workflow_testing_indicator"

file_loc = lambda chapter_num, nb_name: f"[{nb_name}]({repo_root}/{chapter_num}/{nb_name})"
is_run = (
    lambda chapter_num, nb_name: f'<img width="20" alt="image" src={run_root}/notebooks/book1/{chapter_num}/{nb_name}>'
)
get_log = lambda chapter_num, nb_name: f"[log]({run_root}/notebooks/book1/{chapter_num}/{nb_name})"

base_string = lambda chapter: f"<details>\n" f"<summary>Chapter: {chapter}</summary>\n"

for chapter, info in sorted(metadata.items()):
    base_df = pd.DataFrame(columns=["fig", "notebook", "workflow", "fine-tune", "latexify", "jaxify", "interactive"])
    chapter_num = chapter.split("_", 1)[0].zfill(2)
    print(base_string(chapter))
    for idx, (fig_num, py_files) in enumerate(sorted(info.items()), 1):
#         print(fig_num, type(fig_num))
        for py_file in py_files:
            raw_info = [
                fig_num,
                file_loc(chapter_num, py_file.replace(".py", ".ipynb")),
                (
                    is_run(chapter_num, py_file.replace(".py", ".png")),
                    get_log(chapter_num, py_file.replace(".py", ".log")),
                ),
            ] + [f"({wrong}, PR)"] * 4
            base_df.loc[idx, :] = raw_info
    print(base_df.to_markdown(index=None, floatfmt=".2f"))
    print("</details>")
    print("\n")
```
